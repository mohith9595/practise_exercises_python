```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
sns.set()
```

```python
df_used_cars = pd.read_csv('bar_chart_data.csv')
df_used_cars
```

```python
plt.figure(figsize = (9, 6))
plt.bar(x = df_used_cars['Brand'], height = df_used_cars['Cars Listings'], color = 'midnightblue')
plt.xticks(rotation = 60, fontsize = 10)
plt.yticks(fontsize = 10)
plt.title('cars listings by brand', fontsize = 13, fontweight = 'bold' )
plt.ylabel('Number of Listings', fontweight = 'bold', fontsize = 13 )
plt.xlabel('Car brand', fontweight = 'bold', fontsize = 13)
plt.show
```

```python
df_fuel_engine_type = pd.read_csv('pie_chart_data.csv')
```

```python
df_fuel_engine_type
```

```python
plt.figure(figsize =(12, 8))
plt.pie(df_fuel_engine_type["Number of Cars"], labels = df_fuel_engine_type["Engine Fuel Type"].values, autopct = '%.1f%%')
textprops = {'fontsize' : 'bold'}
plt.legend()
plt.title('Cars by Engine Type', fontsize = 16, fontweight = 'bold')
plt.show()
```

```python
df_fuel_engine_type_2 = pd.read_csv('stacked_area_chart_data.csv')
df_fuel_engine_type_2
```

```python
labels = ['Diesel', 'Petrol', 'Gas']
plt.figure(figsize = (12, 6))
plt.stackplot(df_fuel_engine_type_2['Year'],df_fuel_engine_type_2['Petrol'],df_fuel_engine_type_2['Diesel'],df_fuel_engine_type_2['Gas'] , edgecolor = 'none')
plt.xticks(df_fuel_engine_type_2['Year'], rotation = 45)
plt.legend(labels = labels, loc = 'upper left')
plt.show()
```

```python
df_snp_ftse_H1 = pd.read_csv('line_chart_data.csv')
df_snp_ftse_H1
```

```python
df_snp_ftse_H1 ["new_date"] = pd.to_datetime(df_snp_ftse_H1["Date"])
```

```python
df_snp_ftse_H1 ["new_date"]
```

```python
labels = ['S&P 500', 'FTSE100']
plt.figure(figsize = (20, 8 ))
plt.plot(df_snp_ftse_H1['new_date'], df_snp_ftse_H1['GSPC500'])
plt.plot(df_snp_ftse_H1['new_date'], df_snp_ftse_H1['FTSE100'])
plt.title('S&P vs FTSE100 Returns(2000-2010)', fontsize = 15 , fontweight = 'bold')
plt.ylabel = 'Returns'
plt.xlabel = 'Date'
plt.legend(labels = labels, fontsize = 'large')
plt.show()
```

```python
df_snp_ftse_H2 = df_snp_ftse_H1[(df_snp_ftse_H1.new_date >= '2008-07-01') & (df_snp_ftse_H1.new_date <= '2008-12-31')]
```

```python
df_snp_ftse_H2
```

```python
labels = ['S&P 500', 'FTSE100']
plt.figure(figsize = (20, 8 ))
plt.plot(df_snp_ftse_H2['new_date'], df_snp_ftse_H2['GSPC500'], color = 'midnightblue')
plt.plot(df_snp_ftse_H2['new_date'], df_snp_ftse_H2['FTSE100'], color = 'crimson')
plt.title('S&P vs FTSE100 Returns(H2-2008)', fontsize = 15 , fontweight = 'bold')
plt.ylabel('Returns')
plt.xlabel('Date')
plt.legend(labels = labels, fontsize = 'large')
plt.show()
```

```python
df_histogram = pd.read_csv('histogram_data.csv')
df_histogram
```

```python
plt.figure(figsize = (10, 6))
plt.hist(df_histogram['Price'], bins = 10, color = 'crimson')
plt.xlabel('dollars')
plt.ylabel('number of properties')
plt.title('Distribution of Real estate prices')
plt.show()
```

```python
df_real_estate = pd.read_csv('scatter_data.csv')
df_real_estate
```

```python
plt.figure(figsize = (12, 6))
scatter = plt.scatter(df_real_estate['Area (ft.)'], df_real_estate['Price'],
            alpha = 0.8,
            c = df_real_estate['Building Type'],
            cmap = 'viridis')
plt.legend(*scatter.legend_elements(), loc = 'upper left', title = 'Building Type')
plt.show()
```
